```python
df = pd.read_csv('Data Analysis Project - Raw.csv')
df.head()
```

```python
# 1. Create the list of distance columns
distance_columns = [
    'TotalDistance', 
    'TrackerDistance', 
    'LoggedActivitiesDistance', 
    'VeryActiveDistance', 
    'ModeratelyActiveDistance', 
    'LightActiveDistance', 
    'SedentaryActiveDistance'
]

# 2. Round the columns directly on your main dataframe ('df')
df[distance_columns] = df[distance_columns].round(2)

# 3. Verify the changes by previewing the columns
df[['Id', 'ActivityDate'] + distance_columns].head()
```

```python
# 1. Ensure the date column is formatted correctly as actual dates
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])

# 2. Filter out rows where active metrics are 0
valid_rows_df = df[
    (df['TotalSteps'] > 0) & 
    (df['TotalDistance'] > 0) & 
    (df['SedentaryMinutes'] > 0)
]

# 3. Restrict rows to the date range: '2016-05-01' AND '2016-05-12'
date_filtered_df = valid_rows_df[
    (valid_rows_df['ActivityDate'] >= '2016-05-01') & 
    (valid_rows_df['ActivityDate'] <= '2016-05-12')
]

# 4. Group by User ID and count unique activity dates (HAVING COUNT(DISTINCT) = 12)
user_date_counts = date_filtered_df.groupby('Id')['ActivityDate'].nunique()
valid_user_ids = user_date_counts[user_date_counts == 12].index

# 5. Extract only the matching users into your final analysis dataset
final_cleaned_df = date_filtered_df[date_filtered_df['Id'].isin(valid_user_ids)]

# 6. Print confirmation and preview the results
print(f"Filtering complete! Found {len(valid_user_ids)} users who tracked all 12 days.")
final_cleaned_df.head()
```

```python
# Reset the index numbers to start from 0 sequentially
final_cleaned_df = final_cleaned_df.reset_index(drop=True)

# View the freshly numbered dataset
final_cleaned_df
```
